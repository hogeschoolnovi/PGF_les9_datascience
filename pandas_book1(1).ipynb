## Conditionele subsets (filters) 

Stel ik wil alle passagiers boven de leeftijd van 35, dan krijg ik een subset van 217

```python
above_35 = titanic[titanic["Age"] > 35]
above_35
```

Stel ik wil een subset van alle passagiers die in cabine hadden in klasse 2 of 3 

We gebruiken `isin(array_met_waardes)` om te checken of het aan een van de meegegeven waardes voldoet.

```python
class_23 = titanic[titanic["Pclass"].isin([2, 3])]
```

```python
class_23
```

Je kunt hetzelfde resultaat bereiken met een `|` (or character) er tussen

```python
class_23 = titanic[(titanic["Pclass"] == 2) | (titanic["Pclass"] == 3)]
class_23
```

## Subset met rij en kolom

Met de `loc()` methode kun je zowel een rij als een kolom specificeren 

Bijvoorbeeld als je de namen van alle passagiers boven de 35 wilt hebben

```python
adult_names = titanic.loc[titanic["Age"] > 35, "Name"]
adult_names
```

# Nieuwe kolom
De titanic vaarde ongeveer 107.5 uur voordat het de ijsberg raakte. Laten we een kolom toevoegen die berekend hoeveel elke passagier per uur heeft betaald

```python
titanic["hourly"] = titanic["Fare"] / 107.5 
titanic[["Name", "hourly"]]
```

# Statistieken

```python
titanic[["Sex", "Age"]].groupby("Sex").mean()
```

```python
titanic[["Age", "Fare"]].describe()
```

In plaats van de standaard infor van `describe()` kun je ook je eigen collectie van statistische berekeningen maken met `agg()`

```python
titanic.agg(

    {

        "Age": ["min", "max", "median", "skew"],

        "Fare": ["min", "max", "median", "mean"],

    }

)
```

De `grouBy()` functie is handig om je data logisch te presenteren. 

```python
titanic[["Sex", "Age"]].groupby("Sex").mean()
```

Je kun de totale waardes optellen

```python
titanic.groupby("Survived")["Survived"].count()
```

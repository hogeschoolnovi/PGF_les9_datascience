## 1. Lineaire Algebra met SciPy (`scipy.linalg`)

In dit voorbeeld berekenen we de **eigenwaarden** en **eigenvectoren** van een matrix. Dit zijn fundamentele concepten in de lineaire algebra.

Eigenwaarden geven aan hoeveel een vector wordt uitgerekt door een lineaire transformatie (voorstellingen van matrices), terwijl eigenvectoren de richting aangeven waarin dat gebeurt.

```python
from scipy import linalg
import numpy as np

# Matrix A definieren
A = np.array([[1, 2], [3, 4]])

# Bereken de eigenwaarden en eigenvectoren van de matrix A
eigenwaarden, eigenvectoren = linalg.eig(A)

# Resultaten weergeven
print('Eigenwaarden:', eigenwaarden)
print('Eigenvectoren:', eigenvectoren)
```

## 2. Numerieke Integratie met SciPy (`scipy.integrate`)

Numerieke integratie is het proces van het vinden van het gebied onder een curve wanneer er geen eenvoudige algebraïsche oplossing is. Hier gebruiken we SciPy om een eenvoudige functie over een interval te integreren.

```python
from scipy import integrate

# Definieer de functie f(x) = x^2
f = lambda x: x**2

# Bereken de integratie van f(x) over het interval van 0 tot 1
result, error = integrate.quad(f, 0, 1)

# Resultaten weergeven
print('Geïntegreerde waarde:', result)
print('Integratie Error:', error)
```

## 3. Optimalisatie met SciPy (`scipy.optimize`)

Optimalisatie draait om het vinden van minima of maxima van functies. Hier gebruiken we `minimize` om het minimum van een kwadratische functie te zoeken.

```python
from scipy import optimize

# Definieer een kwadratische functie f(x) = (x - 3)^2
f = lambda x: (x - 3)**2

# Vind het minimum van de functie met een startpunt van x = 5
result = optimize.minimize(f, x0=5)

# Resultaten weergeven
print('Minimum gevonden bij x =', result.x[0])
```

## 4. Interpolatie met SciPy (`scipy.interpolate`)

Interpolatie is het proces van het vinden van nieuwe gegevenspunten binnen het bereik van een discrete reeks bekende gegevenspunten. Hier gebruiken we kubische interpolatie om een vloeiende curve te creëren door de gegevenspunten.

```python
from scipy import interpolate
import numpy as np
import matplotlib.pyplot as plt

# Gegevenspunten, bijvoorbeeld een kwadratische functie y = x^2
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([0, 1, 4, 9, 16, 25])

# Creëer een cubische interpolatiefunctie
f = interpolate.interp1d(x, y, kind='cubic')

# Interpoleer nieuwe punten
nieuwe_x = np.linspace(0, 5, 100)
nieuwe_y = f(nieuwe_x)

# Plot de originele punten en de interpolatie
plt.plot(x, y, 'o', label='Originele punten')
plt.plot(nieuwe_x, nieuwe_y, '-', label='Cubische interpolatie')
plt.legend()
plt.show()
```

## 5. Signaalverwerking met SciPy (`scipy.signal`)

In dit voorbeeld gaan we een ruisachtig signaal filteren met een Butterworth laagdoorlaatfilter, wat vooral handig is in de digitale signaalverwerking.

```python
from scipy import signal
import numpy as np
import matplotlib.pyplot as plt

# Tijdreeks voor een sinusvormig signaal met ruis
t = np.linspace(0, 1, 500, endpoint=False)
signaal = np.sin(2 * np.pi * 7 * t) + np.random.randn(500)

# Pas een Butterworth low-pass filter toe (Laagdoorlaatfilter)
b, a = signal.butter(3, 0.05)
gefilterd_signaal = signal.filtfilt(b, a, signaal)

# Plot het originele signaal en het gefilterde signaal
plt.plot(t, signaal, label='Origin. signaal')
plt.plot(t, gefilterd_signaal, label='Gefilterd signaal')
plt.legend()
plt.show()
```

## 6. Statistiek met SciPy (`scipy.stats`)

SciPy biedt uitgebreide statistische functionaliteit, waaronder het uitvoeren van een **t-test** (bepalen of er een significant verschil is). In dit voorbeeld voeren we een **t-test** uit op gegevens die volgen uit een normale verdeling.

```python
from scipy import stats
import numpy as np

# Trek 100 willekeurige getallen uit een normale verdeling
data = np.random.normal(loc=0, scale=1, size=100)

# Bereken de gemiddelde en standaarddeviatie
mean, stddev = np.mean(data), np.std(data)
print(f"Gemiddelde: {mean}, Standaarddeviatie: {stddev}")

# Voer een t-test uit voor een sample met gemiddeld 0
t_stat, p_value = stats.ttest_1samp(data, 0)

# Resultaten weergeven
print(f"T-test statistiek: {t_stat}, P-waarde: {p_value}")
```

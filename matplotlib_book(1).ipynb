```python
import matplotlib.pyplot as plt
import numpy as np
```

# Lijnplot 
    plot()

Lijnplotten zijn een van de meest gebruikte plots. Ze verbinden de punten in je gegevens met lijnen, wat handig is voor het visualiseren van trends.
     

```python
# Gegevens
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

# Maak een lijnplot
plt.plot(x, y)
plt.xlabel('x-as')
plt.ylabel('y-as')
plt.title('Lijnplot')
plt.show()
```

# Scatter diagram
    scatter()

Een scatterplot (spreidingsdiagram) is goed voor het weergeven van de relatie tussen twee variabelen. Elk datapunt wordt weergegeven door een punt of symbool.

```python
# Gegevens
x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

# Maak een scatter plot
plt.scatter(x, y)
plt.xlabel('x-as')
plt.ylabel('y-as')
plt.title('Spreidingsdiagram')
plt.show()
```

Of je maakt het heel gecompliceerd

```python
# Fixing random state for reproducibility
np.random.seed(19680801)


N = 100
r0 = 0.6
x = 0.9 * np.random.rand(N)
y = 0.9 * np.random.rand(N)
area = (20 * np.random.rand(N))**2  # 0 to 10 point radii
c = np.sqrt(area)
r = np.sqrt(x ** 2 + y ** 2)
area1 = np.ma.masked_where(r < r0, area)
area2 = np.ma.masked_where(r >= r0, area)
plt.scatter(x, y, s=area1, marker='^', c=c)
plt.scatter(x, y, s=area2, marker='o', c=c)
# Show the boundary between the regions:
theta = np.arange(0, np.pi / 2, 0.01)
plt.plot(r0 * np.cos(theta), r0 * np.sin(theta))

plt.show()
```

# Staafdiagram
    bar()

Staafdiagrammen zijn handig om discrete categorien te vergelijken. De hoogte van elk staafje komt overeen met de waarde van de categorie.

```python
import matplotlib.pyplot as plt

# Gegevens
categorien = ['A', 'B', 'C', 'D']
waarden = [5, 7, 3, 8]

# Maak een simpel staafdiagram
plt.bar(categorien, waarden)
plt.xlabel('Categorien')
plt.ylabel('Waarde')
plt.title('Staafdiagram')
plt.show()
```

Je kunt ze ook stapelen en labelen

```python
species = ('Adelie', 'Chinstrap', 'Gentoo')
sex_counts = {
    'Male': np.array([73, 34, 61]),
    'Female': np.array([73, 34, 58]),
}
width = 0.6  # the width of the bars: can also be len(x) sequence


fig, ax = plt.subplots()
bottom = np.zeros(3)

for sex, sex_count in sex_counts.items():
    p = ax.bar(species, sex_count, width, label=sex, bottom=bottom)
    bottom += sex_count

    ax.bar_label(p, label_type='center')

ax.set_title('Number of penguins by sex')
ax.legend()

plt.show()
```

# Histogram
    hist()
Een histogram is handig om de verdeling van een dataset te visualiseren. Het verdeelt de data in intervallen en laat het aantal datapunten in elk interval zien.

```python
# Random data genereren
data = np.random.randn(1000)

# Maak het histogram
plt.hist(data, bins=30)
plt.xlabel('Waardes')
plt.ylabel('Frequentie')
plt.title('Histogram')
plt.show()
```

# Boxplot
    box()
Een boxplot wordt gebruikt om de verdeling van data weer te geven en biedt een overzicht van de vijf belangrijkste kwantielen (minimum, eerste kwartiel, mediaan, derde kwartiel en maximum), inclusief eventuele uitschieters (outliers).
     

```python
import matplotlib.pyplot as plt
import numpy as np

# Random data genereren
data = [np.random.normal(0, std, 100) for std in range(1, 4)]

# Boxplot
plt.boxplot(data)
plt.xlabel('Groepen')
plt.ylabel('Waardes')
plt.title('Boxplot')
plt.show()
```

# Subplots
    subplots()
Soms wil je meerdere plots naast elkaar zetten in n figuur. Hier komt de subplots() functie van pas. Het laat je meerdere assen in dezelfde figuur plaatsen.

```python
# Maak subplots met 1 rij en 2 kolommen
fig, axs = plt.subplots(1, 2)

# Linker plot
axs[0].plot([1, 2, 3], [1, 4, 9])
axs[0].set_title('Plot 1')

# Rechter plot
axs[1].plot([1,2,3,4,5], [2, 5, 7, 9, 10])
axs[1].set_title('Plot 2')

# Laat de plots zien
plt.tight_layout()
plt.show()
```
